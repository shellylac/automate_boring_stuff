```python
import random

def guess_game():
    computer_number  = random.randint(1,20)
    guess = int(input("Guess a number between 1 and 20?"))
    n_guesses = 1

    while guess != computer_number and n_guesses < 6:
        if guess < computer_number:
            print("Sorry, too low. Guess again!")
            guess = int(input())
            n_guesses = n_guesses + 1
        elif guess > computer_number:
            print("Sorry, too high. Guess again!")
            guess = int(input())
            n_guesses = n_guesses + 1
        else:
            break
    if guess == computer_number:
        print('You guessed my number in ' + str(n_guesses) + ' guesses!')
    else:
        print('Nope. The number I was thinking of was ' + str(computer_number))
    

guess_game()
```

```python
# This is a guess the number game.
import random

# using a for loop
def guess_game():
    secretNumber = random.randint(1, 20)
    print('I am thinking of a number between 1 and 20.')

    # Ask the player to guess 5 times.
    for guessesTaken in range(1, 6):
        print('Take a guess.')
        guess = int(input())

        if guess < secretNumber:
            print('Your guess is too low.')
        elif guess > secretNumber:
            print('Your guess is too high.')
        else:
            break    # This condition is the correct guess!

    if guess == secretNumber:
        print('Good job! You guessed my number in ' + str(guessesTaken) + 'guesses!')
    else:
        print('Nope. The number I was thinking of was ' + str(secretNumber))
        
        
guess_game()
```

```python
import random, sys

def rps_game():
    print("Let's play 'ROCK, PAPER, SCISSORS!'")
    wins = 0
    losses = 0
    ties = 0

    print("Enter your move: (r)ock (p)aper (s)cissors or (q)uit: ")
    user_choice = input()
    computer_choice = random.choice(['r', 'p', 's'])
        
    if user_choice == 'q':
        sys.exit()
    
    if user_choice not in ['r', 'p', 's', 'q']:
        print("Incorrect symbol entered.")
        print("Enter your move: (r)ock (p)aper (s)cissors or (q)uit: ")
        user_choice = input()
        
    if (user_choice == 'r' and computer_choice == 's') or (user_choice == 'p' and computer_choice == 'r') or  (user_choice == 's' and computer_choice == 'p'):
        print(f"{user_choice} VS {computer_choice} -> You win!" )
        wins = wins + 1
        print('%s Wins, %s Losses, %s Ties' % (wins, losses, ties))
    elif (user_choice == 'p' and computer_choice == 's') or (user_choice == 's' and computer_choice == 'r') or  (user_choice == 'r' and computer_choice == 'p'):
        print(f"{user_choice} VS {computer_choice} -> You lose!")
        losses = losses + 1
        print('%s Wins, %s Losses, %s Ties' % (wins, losses, ties))
    else:
        print(f"{user_choice} VS {computer_choice} -> It's a draw")
        ties = ties + 1
        print('%s Wins, %s Losses, %s Ties' % (wins, losses, ties))
    print("Best of 5 games:" + str(5 - games) + " games left.")
    

rps_game()
```

```python
def collatz_recurring(number):
    while number != 1:
        if number % 2 == 0:
            number = number // 2
            print(number)
        else:
            number = 3 * number + 1
            print(number)
    return(number)

collatz_recurring(56)
 
```
